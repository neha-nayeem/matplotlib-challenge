## Observations and Insights 

&#9656;  There appears to be an almost equal distribution of female to male mice in the overall study, with female mice at **49.6%** and male mice at **50.4%**. <br>

&#9656;  From the box plot and quartile calculations of the four promising drug treatments, it is shown that the only drug regimen to have any outliers is Infubinol, with only one data point outside the lower boundary of **36.83**. This shows that *the statistical distribution of data is normal*. <br>

&#9656;  The box plot also indicates that the mean and median values, as well as the interquartile ranges and boundaries of the Capomulin and Ramicane treatments are lower than those of Infubinol and Ceftamin. These results suggest that Capomulin and Ramicane were more successful in treating tumors and appear to have had the best results, with the final tumor volumes being much lower than the other two. <br>

&#9656;  Capomulin's positive effects are further shown in the line plot where we see the tumor volume for mouse test subject 's185' reduced by almost half at the end of the study. <br>

&#9656;  The scatterplot and linear regression model shows us a strong positive correlation **(0.84)** between average tumor volumes and mouse weights in the Capomulin treatment. This indicates that on average, the bigger the mouse, the larger the tumor. <br>

```python
# --- dependencies and setup ---
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# --- get the file path of CSV files ---
mouse_data_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# --- read the mouse data and the study results into dataframes using Pandas ---
mouse_data = pd.read_csv(mouse_data_path)
study_results = pd.read_csv(study_results_path)

# --- combine the data into a single dataset on Mouse ID column ---
merged_df = pd.merge(mouse_data, study_results, on="Mouse ID")

# --- display the merged dataframe for preview ---
merged_df.head()
```

```python
# ****************************
# Checking the number of mice.
# ****************************

TotalMice = len(merged_df["Mouse ID"])

# --- find the number of unique mice in the DataFrame ---
UniqueMice = len(merged_df["Mouse ID"].unique())

# --- create a dataframe to display mice numbers --- 
MiceNumbersDF = pd.DataFrame({"Total Mice": [TotalMice],
                              "Unique Mice": [UniqueMice]})
MiceNumbersDF
```

```python
# *********************************************************************************
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# *********************************************************************************

DuplicateMouseData = merged_df[merged_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]

# --- display all the rows that have duplicate Mouse ID AND Timepoint ---
DuplicateMouseData
```

```python
# ******************************************************
# Optional: Get all the data for the duplicate mouse ID. 
# ******************************************************

# --- retrieve all the rows for Mouse ID g989 found above ---
DuplicateMouseData = merged_df.loc[merged_df["Mouse ID"] == "g989", :]
DuplicateMouseData
```

```python
# *******************************************************************
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# *******************************************************************

# --- get indexes of the duplicate mouse ID g989 rows ---
DuplicateMouseIndex = DuplicateMouseData.index

# --- create a clean DataFrame by dropping the duplicate mouse by its indice found above ---
cleanDF = merged_df.drop(DuplicateMouseIndex).reset_index(drop = True)
cleanDF.head()
```

```python
# ***************************************************
# Checking the number of mice in the clean DataFrame.
# ***************************************************

TotalMice = len(cleanDF["Mouse ID"])

# --- find the number of unique mice in the clean dataFrame ---
UniqueMice = len(cleanDF["Mouse ID"].unique())

# --- create a dataframe to display mice numbers --- 
MiceNumbersCleanedDF = pd.DataFrame({"Total Mice": [TotalMice],
                              "Unique Mice": [UniqueMice]})
MiceNumbersCleanedDF
```

## Summary Statistics

```python
# *******************************************************************************************************************************
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
# *******************************************************************************************************************************

# --- create a dataframe groupby object on drug regimen ---
DrugGroups = cleanDF.groupby(["Drug Regimen"])

# --- calculate tumor volume statistics on the groupby object and store as a series ---
TumorVolMean = DrugGroups["Tumor Volume (mm3)"].mean()
TumorVolMedian = DrugGroups["Tumor Volume (mm3)"].median()
TumorVolVar = DrugGroups["Tumor Volume (mm3)"].var()
TumorVolStd = DrugGroups["Tumor Volume (mm3)"].std()
TumorVolSem = DrugGroups["Tumor Volume (mm3)"].sem()

# --- create a new dataframe with the calculated statistics above and display the table ---
SummaryStatsDF = pd.DataFrame({"Tumor Volume Mean (mm3)": round(TumorVolMean, 2),
                               "Tumor Volume Median (mm3)": round(TumorVolMedian, 2),
                               "Tumor Volume Variance (mm3)": round(TumorVolVar, 2),
                               "Tumor Volume Standard Deviation": round(TumorVolStd, 2),
                               "Tumor Volume SEM": round(TumorVolSem, 2)})

SummaryStatsDF
```

```python
# ********************************************************************************************************************************
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
# ********************************************************************************************************************************

# --- use .groupby().agg() to accomplish the summary statistics table ---
SummaryGroupStats = pd.DataFrame(cleanDF.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"]).round(2))
SummaryGroupStats
```

## Quartiles, Outliers and Boxplots

```python
# *************************************************************************************
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# *************************************************************************************

# --- create a dictionary with promising treatments as keys and empty lists as values --- 
# --- (keys for filtering and labelling & values to be populated in the for loop later) ---
TumorVolumeByTreatment = {"Capomulin": [], "Ramicane": [], "Infubinol": [], "Ceftamin": []}

# --- retrieve the rows of the four drug regimens above into a new dataframe ---
TreatmentsFinalTimepointDF = cleanDF.loc[cleanDF["Drug Regimen"].isin(TumorVolumeByTreatment.keys()), :].copy()

# --- find the final tumor value by sorting items and dropping duplicates except the last one (final timepoint) ---
TreatmentsFinalTimepointDF.sort_values(by=["Mouse ID", "Timepoint"], inplace=True)
TreatmentsFinalTimepointDF.drop_duplicates("Mouse ID", keep="last", inplace=True)
TreatmentsFinalTimepointDF.sort_values(by=["Drug Regimen"], inplace=True)
TreatmentsFinalTimepointDF.reset_index(drop=True, inplace=True)

# --- display final timepoint dataframe ---
TreatmentsFinalTimepointDF.head()

```

```python
# ***********************************************************************************
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# ***********************************************************************************
    
# --- for loop to populate dictionary values with final tumor volumes, calculate and display quartiles/outlier info for each treatment ---
for key, value in TumorVolumeByTreatment.items():

    # --- locate the rows which contain mice on each drug and get the tumor volumes ---
    value = TreatmentsFinalTimepointDF.loc[TreatmentsFinalTimepointDF["Drug Regimen"] == key, "Tumor Volume (mm3)"]
    TumorVolumeByTreatment[key].append(value)

    # --- calculate the IQR and quantitatively determine if there are any potential outliers --- 
    Quartiles = value.quantile([0.25, 0.5, 0.75])
    LowerQuart = Quartiles[0.25]
    MedianQuart = Quartiles[0.5]
    UpperQuart = Quartiles[0.75]
    IQR = UpperQuart - LowerQuart
    
    # --- determine outliers using upper and lower bounds ---
    Lowerbound = LowerQuart - (1.5*IQR)
    Upperbound = UpperQuart + (1.5*IQR)


    # --- display quartile and outlier info across the four drug regimens ---
    print(f"-------------------------------------------------")   
    print(f"Final Tumor Volume Outliers - {key}")
    print(f"-------------------------------------------------")
    print(f"First Quartile: {round(LowerQuart, 2)}")
    print(f"Second Quartile: {round(MedianQuart, 2)}")
    print(f"Third Quartile: {round(UpperQuart, 2)}")
    print(f"Interquartile Range (IQR): {round(IQR, 2)}")
    print(f"Values below {round(Lowerbound, 2)} could be potential outliers.")
    print(f"Values above {round(Upperbound, 2)} could be potential outliers.\n")

```

```python
# ********************************************************************************************
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# ********************************************************************************************

# --- create a list with dictionary values for plotting and retrieve labels from dictionary keys ---
PlotValues = [TumorVolumeByTreatment["Capomulin"][0], TumorVolumeByTreatment["Ramicane"][0], 
              TumorVolumeByTreatment["Infubinol"][0], TumorVolumeByTreatment["Ceftamin"][0]]
labels = TumorVolumeByTreatment.keys()

# --- create dictionaries to style the boxplot elements ---
flierprops = {"marker": "o", "markerfacecolor": "darksalmon", "markersize": 8, "linestyle": "none", 
              "markeredgecolor": "darksalmon"}
medianprops = {"color" : "limegreen"}
meanprops = {"marker": "s", "markerfacecolor": "steelblue", "markeredgecolor": "steelblue"}

# --- create a boxplot showing the results ---
fig, ax = plt.subplots(figsize=(8, 6))
ax.boxplot(PlotValues, labels=labels, showmeans=True, meanprops=meanprops, flierprops=flierprops, 
           medianprops=medianprops)

# --- set title and label ---
ax.set_title("Comparison of Final Tumor Volumes across Drug Regimens of Interest", fontsize=14)
ax.set_ylabel("Final Tumor Volume (mm3)", fontsize=12)
ax.set_xlabel("Drug Regimen", fontsize=12)

plt.tight_layout()
plt.show()

```

## Line and Scatter Plots

```python
# *****************************************************************************************
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
# *****************************************************************************************

# --- retrieve mouse ID s185 treated for Capomulin from the dataframe ---
Cap_mouse = cleanDF.loc[cleanDF["Mouse ID"] == "s185", :].reset_index(drop=True)

# --- get the timepoint and tumor volume data ---
Timepoint = Cap_mouse["Timepoint"]
TumVol = Cap_mouse["Tumor Volume (mm3)"]

# --- plot a line graph of timepoint vs tumor volume ---
plt.figure(figsize=(10,6))
plt.plot(Timepoint, TumVol, color="lightseagreen", linewidth=2, marker="s", markersize=7)

# --- set title, labels, limits --- 
plt.xlabel("Timepoint (in days)", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontsize=12)
plt.title("Tumor Volume (mm3) over Time for Mouse ID s185 (on Capomulin)", fontsize=14)
plt.ylim(20,max(TumVol)+3)
plt.xlim(0, max(Timepoint))

plt.tight_layout()
plt.show()

```

```python
# *********************************************************************************************
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
# *********************************************************************************************

# --- retrieve all the rows for mice treated with Capomulin and create a groupby mean object ---
Capomulin_data = cleanDF.loc[cleanDF["Drug Regimen"] == "Capomulin", :].reset_index(drop=True)
MeanCap = Capomulin_data.groupby("Mouse ID").mean()

# --- retrieve mouse weights and average tumor volumes ---
MouseWeights = MeanCap["Weight (g)"]
AvgTumorVol = MeanCap["Tumor Volume (mm3)"]

# --- create a scatterplot with above values ---
plt.figure(figsize=(10,6))
plt.scatter(MouseWeights, AvgTumorVol, marker="o", s=50, facecolors="lightseagreen", edgecolors="black", alpha=0.75)

# --- set title, labels, limits ---
plt.title ("Capomulin Regimen: Average Tumor Volume vs Mouse Weight", fontsize = 14)
plt.xlabel("Mouse Weight (g)", fontsize=12)
plt.ylabel("Average Tumor Volume (mm3)", fontsize=12)
plt.xlim(min(MouseWeights)-0.5, max(MouseWeights)+0.5)
plt.ylim(min(AvgTumorVol)-1, max(AvgTumorVol)+1)
plt.show()
```

## Correlation and Regression

```python
# ********************************************************************
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# ********************************************************************

# --- Define x and y for the line equation ---
x = MouseWeights
y = AvgTumorVol

# y = m * x + b
# AvgTumor Volume = slope * MouseWeights + intercept
# --- calculate linear regression values using scipy linregress() ---
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)

# --- calculate correlation coefficient using scipy pearsonr() --- 
coefficient = round(st.pearsonr(x,y)[0], 2)

# --- display calculated values of linear regression and correlation coefficients ---
print(f"------------------------------------------")
print(f"Linear Regression Values")
print(f"------------------------------------------")
print(f"Slope: {round(slope, 2)}")
print(f"Intercept: {round(intercept, 2)}")
print(f"R Value: {round(rvalue, 2)}")
print(f"Pearson R Correlation Coefficient: {coefficient}")
print(f"Std Error: {round(stderr, 2)}")
print(f"R Squared: {round(rvalue**2, 2)}\n")
print(f"------------------------------------------")

# --- add the linear regression equation and line to scatter plot ---
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))

plt.figure(figsize=(10,6))
plt.scatter(x, y, marker="o", s=50, facecolors="lightseagreen", edgecolors="black", alpha=0.75)
plt.plot(x, regress_values, "darksalmon")
plt.annotate(line_eq,(21, 37),fontsize=20,color="indianred")

# --- set title, labels, limits ---
plt.title ("Capomulin Regimen: Average Tumor Volume vs Mouse Weight", fontsize = 14)
plt.xlabel("Mouse Weight (g)", fontsize=12)
plt.ylabel("Average Tumor Volume (mm3)", fontsize=12)
plt.xlim(min(MouseWeights)-0.5, max(MouseWeights)+0.5)
plt.ylim(min(AvgTumorVol)-1, max(AvgTumorVol)+1)

plt.show()
```
